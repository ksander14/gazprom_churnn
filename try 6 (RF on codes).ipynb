```python
import pandas as pd
from datetime import datetime
import tqdm
```

```python
df = pd.read_csv('features_codes_38.csv')
df['date'] = df['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
df.head()
```

```python
train_target = pd.read_csv('train_target.csv')
train_target['date'] = train_target['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
train_target.head()
```

```python
X = df.set_index(['id', 'date']).join(train_target.set_index(['id', 'date']))
X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
clf = RandomForestClassifier(max_depth=4, random_state=0)
clf.fit(X.iloc[:,:-1], X.iloc[:,-1])
```

```python
cross_val_score(clf, X.iloc[:,:-1], X.iloc[:,-1], scoring='roc_auc', cv = 3).mean()
```
