```python
import pandas as pd
from datetime import datetime
import tqdm
```

```python
df_train = pd.read_csv('train_data.csv')
df_train.drop('Unnamed: 0', axis=1, inplace=True)
df_train['date'] = df_train['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d').replace(day = 1))
df_train.head()
```

```python
df_train = df_train[(df_train.date >= datetime(2017,6,1)) & (df_train.date < datetime(2017,12,1))]
df_train.head()
```

```python
df_train['code'] = df_train['code'].apply(lambda x: str(x)[1:] if str(x)[0] == '0' else str(x))
df_train['code'] = df_train['code'].apply(lambda x: '999' if '?' in str(x) else str(x))
```

```python
df_train['code0'] = df_train['code'].apply(lambda x: str(x)[:3])
```

```python
print(df_train['code'].nunique())
top_codes = df_train['code'].value_counts()
top_codes.head(40)
```

```python
print(df_train['code0'].nunique())
top_codes = df_train['code0'].value_counts()
top_codes.head(40)
```

```python
df_codes = df_train.groupby(['id', 'date'])['code0'].value_counts()
```

```python
df_codes = pd.DataFrame(df_codes)
df_codes['id'] = [x[0] for x in df_codes.index]
df_codes['date'] = [x[1] for x in df_codes.index]
df_codes['cod'] = [x[2] for x in df_codes.index]
df_codes.head(10)
```

```python
table = pd.pivot_table(df_codes, values='code0', index=['id', 'date'], columns=['cod'], aggfunc=sum)
table = table.fillna(0)
table.head(10)
```

```python
table.to_csv('features_codes_38_Dec.csv')
```

```python
df_train = pd.read_csv('test_data.csv')
df_train.drop('Unnamed: 0', axis=1, inplace=True)
df_train['date'] = df_train['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
df_train.head()
```

```python
df_train['code'] = df_train['code'].apply(lambda x: str(x)[1:] if str(x)[0] == '0' else str(x))
df_train['code'] = df_train['code'].apply(lambda x: '999' if '?' in str(x) else str(x))
df_train['code0'] = df_train['code'].apply(lambda x: str(x)[:3])
top_codes = df_train['code0'].value_counts()
```

```python
df_codes = df_train.groupby(['id'])['code0'].value_counts()
```

```python
df_codes = pd.DataFrame(df_codes)
df_codes['id'] = [x[0] for x in df_codes.index]
df_codes['cod'] = [x[1] for x in df_codes.index]
df_codes.head(10)
```

```python
table = pd.pivot_table(df_codes, values='code0', index='id', columns=['cod'], aggfunc=sum)
table = table.fillna(0)
table.head(10)
```
