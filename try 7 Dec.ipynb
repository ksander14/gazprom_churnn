```python
import pandas as pd
from datetime import datetime
import tqdm
```

```python
df = pd.read_csv('features_codes_38_Dec.csv')
df.head()
```

```python
olya_simple_features_Dec = pd.read_csv('olya_simple_features_Dec.csv')
olya_simple_features_Dec.head()
```

```python
ksusha_features =  pd.read_csv()
```

```python
X = df.set_index('id').join(olya_simple_features_Dec.set_index('id')).join(ksusha_features.set_index('id'))
X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
clf = RandomForestClassifier(max_depth=4, random_state=0)
clf.fit(X.drop('target'), X['target'])
```

```python
cross_val_score(clf, X.iloc[:,:-1], X.iloc[:,-1], scoring='roc_auc', cv = 3).mean()
```

```python
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X.drop('target', axis=1))
scaler.transform(X.drop('target', axis=1))
```

```python
logreg = linear_model.LogisticRegression(C=1e5, n_jobs=10)
```

```python
cross_val_score(logreg, X.drop('target', axis=1), X['target'], scoring='roc_auc', cv = 3).mean()
```
