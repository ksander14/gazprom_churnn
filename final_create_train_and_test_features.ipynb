## Active months

```python
def active_months(transactions, start_date, end_date):
    '''
    start_date and end_date - date format, [start_date, end_date]
    Need column month! - '%Y-%m' format
    return dict {user: active_months}
    if element not in - 0
    example of using result:
    df['active_months'] = df['id'].apply(lambda x: user_month_count_dict.get(x, 0))
    
    '''
    user_month_count_dict = transactions[(transactions['datetime'] >= start_date) & (transactions['datetime'] <= end_date)].groupby(['id'])['month'].nunique()
    return user_month_count_dict
```

```python

def next_month(current_month):
    return (datetime.strptime(current_month, '%Y-%m') + relativedelta.relativedelta(months=1)).strftime('%Y-%m')

def prev_month(current_month):
    return (datetime.strptime(current_month, '%Y-%m') - relativedelta.relativedelta(months=1)).strftime('%Y-%m')
```
