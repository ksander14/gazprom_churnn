```python
import pandas as pd
from datetime import datetime
import tqdm
```

```python
df_train = pd.read_csv('train_data.csv')
df_train.drop('Unnamed: 0', axis=1, inplace=True)
df_train.head()
```

```python
df_train['date'] = df_train['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d').replace(day = 1))
```

```python
df_train.head()
```

```python
def next_month(x):
    if x.month == 12:
        return x.replace(year=x.year+1, month=1)
    else:
        return x.replace(month=x.month+1)
```

```python
for i in tqdm.tqdm(df_train.index):
    if df_train.loc[i, 'id'] in list(df_train[df_train.date == next_month(df_train.loc[0,'date'])]['id']):
        df_train.loc[i, 'target'] = 1
    else:
        df_train.loc[i, 'target'] = 0
```

```python
df_train.head()
```

```python
df_small = df_train[['date', 'id']].drop_duplicates()
df_small.head()
```

```python
for i in tqdm.tqdm(df_small.index):
    if len(df_small[(df_small.id == df_small.loc[i, 'id']) & (df_small.date == next_month(df_small.loc[0,'date']))]) > 0:
        df_small.loc[i, 'target'] = 1
    else:
        df_small.loc[i, 'target'] = 0
```

```python
df_sorted = df_small.sort_values(['id', 'date'])
df_sorted.head(20)
```

```python
df_sorted = df_sorted.reset_index(drop=True)
df_sorted.drop('terget', axis=1, inplace=True)
df_sorted.head(20)
```

```python
for i in tqdm.tqdm(df_sorted.index[:1]):
    if (df_sorted.loc[i, 'id'] == df_sorted.loc[i+1, 'id']) & (df_sorted.loc[i, 'date'] == next_month(df_sorted.loc[i, 'date'])):
        df_sorted.loc[i, 'target'] = 1
    else:
        df_sorted.loc[i, 'target'] = 0
```
