```python
import pandas as pd
from datetime import datetime
from datetime import timedelta
import tqdm
```

```python
df_train = pd.read_csv('test_data.csv')
df_train.drop('Unnamed: 0', axis=1, inplace=True)
df_train['date'] = df_train['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
df_train.head()
```

```python
df_train.shape
```

```python
def split_test(df):
    tmp = df.groupby('id')['date'].max() 
    id_Oct = tmp[tmp < datetime(2017,11,1)].index
    id_Dec = tmp[tmp >= datetime(2017,11,1)].index
    df_Oct = df[df.id.isin(id_Oct)]
    df_Dec = df[df.id.isin(id_Dec)]
    return df_Oct, df_Dec
```

```python
df_Oct, df_Dec = split_test(df_train)
```
