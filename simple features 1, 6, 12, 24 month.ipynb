```python
import pandas as pd
import datetime as dt
```

```python
df_train = pd.read_csv('train_data.csv')
df_train.drop('Unnamed: 0', axis=1, inplace=True)
df_train['date'] = df_train['date'].apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%d'))
```

```python
def get_time_sample(df, month, window):
    end_date = dt.datetime(2017, month-1, 30)
    start_date = end_date - dt.timedelta(int(30.5*window))
    return df[(df.date >= start_date) & (df.date <= end_date)]

def join_df(df1, df2):
     df1.set_index('id').join(df2.set_index('id'))
```

```python
cols = ['q', 'v_l', 'sum_b', 'percent']
```

```python
# фичи для 1 месяца
month = 12
window = 1
df = get_time_sample(df_train, month, window)
features_1 = df.groupby('id')['q', 'v_l', 'sum_b', 'percent'].sum()

col_names = []
for i in cols:
    col_names.append(i + '_' + str(window))
features_1.columns = col_names
features_1.head()
```

```python
# фичи для 12 месяцев
month = 12
window = 12
df = get_time_sample(df_train, month, window)
features_12 = df.groupby('id')['q', 'v_l', 'sum_b', 'percent'].sum()

col_names = []
for i in cols:
    col_names.append(i + '_' + str(window))
features_12.columns = col_names
features_12.head()
```

```python
# фичи для 18 месяцев
month = 12
window = 18
df = get_time_sample(df_train, month, window)
features_18 = df.groupby('id')['q', 'v_l', 'sum_b', 'percent'].sum()

col_names = []
for i in cols:
    col_names.append(i + '_' + str(window))
features_18.columns = col_names
features_18.head()
```

```python
# фичи для 24 месяцев
month = 12
window = 24
df = get_time_sample(df_train, month, window)
features_24 = df.groupby('id')['q', 'v_l', 'sum_b', 'percent'].sum()

col_names = []
for i in cols:
    col_names.append(i + '_' + str(window))
features_24.columns = col_names
features_24.head()
```

```python
result = features_1.join(features_6).join(features_12).join(features_18).join(features_24)
result.head()
```

```python
df_test = pd.read_csv('test_data.csv')
df_test.drop('Unnamed: 0', axis=1, inplace=True)
df_test['date'] = df_test['date'].apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%d'))
df_test.head()
```

```python
df_test['id'].nunique()
```

```python
def split_test(df):
    tmp = df.groupby('id')['date'].max() 
    id_Oct = tmp[tmp < dt.datetime(2017,11,1)].index
    id_Dec = tmp[tmp >= dt.datetime(2017,11,1)].index
    df_Oct = df[df.id.isin(id_Oct)]
    df_Dec = df[df.id.isin(id_Dec)]
    return df_Oct, df_Dec
```

```python
df_Oct, df_Dec = split_test(df_test)
```

```python
df_Oct['id'].nunique() + df_Dec['id'].nunique()
```

```python
def my_features(df, month):
    # фичи для 1 месяца
    window = 1
    df = get_time_sample(df, month, window)
    features_1 = df.groupby('id')['q', 'v_l', 'sum_b', 'percent'].sum()

    col_names = []
    for i in cols:
        col_names.append(i + '_' + str(window))
    features_1.columns = col_names
    features_1.head()

    # фичи для 6 месяцев
    window = 6
    df = get_time_sample(df, month, window)
    features_6 = df.groupby('id')['q', 'v_l', 'sum_b', 'percent'].sum()

    col_names = []
    for i in cols:
        col_names.append(i + '_' + str(window))
    features_6.columns = col_names
    features_6.head()

    # фичи для 12 месяцев
    window = 12
    df = get_time_sample(df, month, window)
    features_12 = df.groupby('id')['q', 'v_l', 'sum_b', 'percent'].sum()

    col_names = []
    for i in cols:
        col_names.append(i + '_' + str(window))
    features_12.columns = col_names
    features_12.head()

    # фичи для 18 месяцев
    window = 18
    df = get_time_sample(df, month, window)
    features_18 = df.groupby('id')['q', 'v_l', 'sum_b', 'percent'].sum()

    col_names = []
    for i in cols:
        col_names.append(i + '_' + str(window))
    features_18.columns = col_names
    features_18.head()

    # фичи для 24 месяцев
    window = 24
    df = get_time_sample(df, month, window)
    features_24 = df.groupby('id')['q', 'v_l', 'sum_b', 'percent'].sum()

    col_names = []
    for i in cols:
        col_names.append(i + '_' + str(window))
    features_24.columns = col_names
    features_24.head()

    return features_1.join(features_6).join(features_12).join(features_18).join(features_24)
```

```python
agg_Oct = my_features(df_Oct, 10)
agg_Dec = my_features(df_Dec, 12)
agg = agg_Oct.append(agg_Dec)
```

```python
agg.to_csv('olya_features_test_all_periods.csv')
```
