```python
import pandas as pd
from datetime import datetime
import tqdm
```

```python
train_target = pd.read_csv('train_target.csv')
train_target['date'] = train_target['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
train_target.head()
```

```python
user_month_azs = pd.read_csv('user_month_time_features.csv')
user_month_azs['id'] = user_month_azs['user_month'].apply(lambda x: x.split('_')[0])
user_month_azs['date'] = user_month_azs['user_month'].apply(lambda x: datetime(2016, int(x.split('_')[1]), 1) if int(x.split('_')[1]) <= 12 else datetime(2017, int(x.split('_')[1])-12, 1))
user_month_azs.drop('user_month', axis=1, inplace=True)
user_month_azs.head()
```

```python
olya_simple_sum_features = pd.read_csv('olya_simple_sum_features.csv')
olya_simple_sum_features['date'] = olya_simple_sum_features['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
olya_simple_sum_features.head()
```

```python
df = pd.read_csv('features_codes_38.csv')
df['date'] = df['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
df.head()
```

```python
X = df.set_index(['id', 'date']).join(user_month_azs.set_index(['id', 'date']))
X = X.join(olya_simple_sum_features.set_index(['id', 'date']))

X = X.join(train_target.set_index(['id', 'date']))
X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
X = X.fillna(0)
```

```python
clf = RandomForestClassifier(max_depth=4, random_state=0, n_jobs=10)
clf.fit(X.drop('target', axis=1), X['target'])
```

```python
cross_val_score(clf, X.drop('target', axis=1), X['target'], scoring='roc_auc', cv = 3).mean()
```

```python
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X.drop('target', axis=1))
scaler.transform(X.drop('target', axis=1))
```

```python
logreg = linear_model.LogisticRegression(C=1e5, n_jobs=10)
```

```python
cross_val_score(logreg, X.drop('target', axis=1), X['target'], scoring='roc_auc', cv = 3).mean()
```
